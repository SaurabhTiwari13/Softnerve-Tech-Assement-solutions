```python
# 1st question solution

def find_leaders(arr):
    n = len(arr)
    leaders = []

    max_right = arr[n - 1]
    leaders.append(max_right)

    for i in range(n - 2, -1, -1):
        if arr[i] > max_right:
            max_right = arr[i]
            leaders.append(max_right)

    leaders.reverse()  
    return leaders


arr = [16, 17, 4, 3, 5, 2]
result = find_leaders(arr)
print("Leaders in the array:", result)
```

```python
# 2nd question solution

def max_profit(prices):
    if not prices or len(prices) < 2 or len(prices) > 105:
        raise ValueError("Invalid length of prices list")

    min_price = prices[0]
    max_profit = 0

    for price in prices[1:]:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)

    return max_profit if max_profit > 0 else 0


prices = [7,1,5,3,6,4]

try:
    result = max_profit(prices)
    print(result)
except ValueError as e:
    print(e)
```

```python
# 3rd question solution

def subsetXORSum(nums):
    def backtrack(start, subset_xor):
        nonlocal total_xor
        total_xor += subset_xor

        for i in range(start, len(nums)):
           
            subset_xor ^= nums[i]
            
            backtrack(i + 1, subset_xor)
            
            subset_xor ^= nums[i]

    total_xor = 0
   
    backtrack(0, 0)
    return total_xor


nums = [1,3]

result = subsetXORSum(nums)
print(result)
```
